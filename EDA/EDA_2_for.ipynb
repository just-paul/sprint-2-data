```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master_df = pd.read_csv('master_df.csv').drop(columns = 'Unnamed: 0')
master_df.head()
```

```python
master_df.columns
```

```python
#####
# foreign artist
audio_feat = ['duration','popularity', 'danceability', 'energy', 'key', 'loudness', 'mode','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

# Audio Features

# Foreign artist 2017

```python
df_for_17 = master_df[(master_df['year'] == 2017) & (master_df['category'] == 'foreign')]
df_for_17.head()
```

```python
for_list17 = df_for_17.groupby(['track_id'])[['streams']].sum().sort_values(by ='streams', ascending = False)[:100]
for_list17 = for_list17.index
for_list17
```

```python
audio_feat_for_df17 = df_for_17[df_for_17['track_id'].isin(for_list17)]
audio_feat_for_df17 = audio_feat_for_df17[audio_feat].drop_duplicates()
audio_feat_for_df17
```

```python
audio_feat_for_df17['duration'] = audio_feat_for_df17['duration'] / 60000
audio_feat_for_df17
```

```python
audio_feat_for_df17.describe()
```
