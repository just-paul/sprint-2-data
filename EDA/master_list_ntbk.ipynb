```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_chart_tracks = pd.read_csv('spotify_daily_charts_tracks.csv')
df_daily_charts = pd.read_csv('spotify_daily_charts.csv')
df_spotify_daily_chart_artists = pd.read_csv('spotify_daily_charts_artists.csv')
```

```python
df_m1 = df_daily_charts.merge(df_chart_tracks, on = 'track_id', how = 'left')
df_final = df_m1.merge(df_spotify_daily_chart_artists, on = 'artist_id', how = 'left')
df_final = df_final.drop(columns = ['track_name_y', 'artist_name_y', 'popularity_y'])
df_final = df_final.rename(columns = {'track_name_x':'track_name','popularity_x':'popularity', 'artist_name_x':'artist_name'})
df_final.columns
```

```python
# to date time conversion
df_final['date'] = pd.to_datetime(df_final['date'])
df_final['year'] = df_final['date'].dt.year
df_final['month'] = df_final['date'].dt.month
df_final['day'] = df_final['date'].dt.day
df_final['day_of_week'] = df_final['date'].dt.dayofweek
```

```python
# list conversion
df_final['genres'] = df_final['genres'].apply(lambda x: x.replace("'", ''))
df_final['genres'] = df_final['genres'].apply(lambda x: x.replace("[", ''))
df_final['genres'] = df_final['genres'].apply(lambda x: x.replace("]", ''))
df_final['genres'] = df_final['genres'].apply(lambda x: x.replace(" ", ''))
df_final['genres'] = df_final['genres'].apply(lambda x: x.split(','))
```

```python
df_final = df_final.fillna(0)
df_final.info()
```

```python
#####
df_genres = df_final.copy()
#####
```

```python
index = 0
lst = []
for item in df_genres['genres']:
    lst.extend(map(lambda x: [index, x], item))
    index += 1
lst[:5]
```

```python
genres = pd.DataFrame(lst, columns=['index', 'genres'])
genres
```

```python
df_genres_final = pd.merge(df_genres.drop('genres', axis = 1), genres, left_index = True, right_on = 'index')
df_genres_final.head()
```

```python
df_genres_final.columns
```

```python
art_opm = df_genres_final[(df_genres_final['genres'] == 'opm')]['artist'].unique()
art_opm
```

```python
# category foreign or opm
df_mod = df_final.copy()
for_mod_df = df_mod[~df_mod['artist'].isin(art_opm)]
for_mod_df['category'] = 'foreign'
for_mod_df
```

```python
opm_mod_df = df_mod[df_mod['artist'].isin(art_opm)]
opm_mod_df['category'] = 'opm'
opm_mod_df
```

```python
labeled_df = pd.concat([for_mod_df, opm_mod_df], axis = 0)
labeled_df.info()
```

```python
labeled_df = labeled_df.dropna()
```

```python
master_df = labeled_df.copy()
master_df.to_csv('master_df.csv')
```
